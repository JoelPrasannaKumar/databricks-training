```python
department_data = [
("IT", "John"),
("HR", "Smith"),
("QA", "David"),
("Analytics", "Kevin"),
("Support", "Robert")
]
dept_columns = ["department", "manager"]
dept_df = spark.createDataFrame(department_data, dept_columns)
dept_df.display()
```

```python
emp_df.join(dept_df,emp_df.department==dept_df.department,"inner").display()
```

```python
emp_df.join(dept_df,emp_df.department==dept_df.department,"left").display()
```

```python
emp_df.join(
    dept_df,
    emp_df.department == dept_df.department,
    "right"
).display()
```

```python
emp_df.join(
    dept_df,
    emp_df.department == dept_df.department,
    "outer"
).display()
```

```python
emp_df.join(
    dept_df,
    emp_df.department == dept_df.department,
    "inner"
).select(
    "name",
    "manager"
).display()
```

### unionByName()

```python
shuffled_df=new_emp_df.select("name","emp_id","department","designation","joining_date","age","city","gender","salary")
emp_df.unionByName(shuffled_df).display()
```

```python
new_emp_df_missing=new_emp_df.drop("gender")
emp_df.unionByName(new_emp_df_missing,allowMissingColumns=True).display()
```

```python
new_emp_df_extra=new_emp_df.withColumn("bonus",lit(5000))
emp_df.unionByName(new_emp_df_extra,allowMissingColumns=True).display()
```

```python
# union does union by matching columns by positions , so order of columns should be same
emp_df.union(new_emp_df).display()

#where as unionByName does union by matching columns by names, so order of columns can be different
emp_df.unionByName(shuffled_df).display()
```

### explode()

```python
skills_data = [
(101, ["Python", "Spark", "Azure"]),
(102, ["Java", "SQL"]),
(103, ["Power BI", "SQL"])
]
skills_columns = ["emp_id", "skills"]
skills_df = spark.createDataFrame(skills_data, skills_columns)
```

```python
skills_df.display()
```

```python
skills_df.select("emp_id",explode("skills").alias("skill")).display()
```

```python
skills_df.select("emp_id",explode("skills").alias("skill")).count()
```

```python
skills_df.filter(array_contains(col("skills"),"Python")).display()
```

```python
skills_df.select(explode(col("skills").alias("skill"))).distinct().display()
```

```python
skills_df.select(explode(col("skills")).alias("skill")).groupBy("skill").count().display()
```

### split()

```python
emp_df.withColumn("joining year",split(col("joining_date"),"-")[0]).\
    withColumn("joining month",split(col("joining_date"),"-")[1]).\
    withColumn("joining day",split(col("joining_date"),"-")[2]).display()
```

```python
emp_df.withColumn(
    "designation_words",
    split(col("designation"), " ")
).display()
```

```python
emp_df.withColumn("city words",split(col("city")," ")).display()
```

```python
emp_df.withColumn(
    "year",
    split(col("joining_date"), "-")[0]
).display()
```

```python
emp_df.withColumn("year", split(col("joining_date"), "-")[0]) \
      .withColumn("month", split(col("joining_date"), "-")[1]) \
      .withColumn("day", split(col("joining_date"), "-")[2]) \
      .display()
```

### concat() / concat_ws() 

```python
emp_df.withColumn("name_department",concat(col("name"),col("department"))).display()
```

```python
emp_df.withColumn("employee_details",concat(col("name"),lit("-"),col("designation"),lit("-"),col("department"))).display()
```

```python
emp_df.withColumn("city_department",concat_ws("-",col("city"),col("department"))).display()
```

```python
emp_df.withColumn(
    "employee_label",
    concat_ws(
        "_",
        col("emp_id"),
        col("name"),
        col("department")
    )
).display()
```

```python
emp_df.withColumn(
    "name_designation",
    concat_ws(
        " - ",
        col("name"),
        col("designation")
    )
).display()
```

### cast()

```python
emp_df.withColumn(
    "age_string",
    col("age").cast("string")
).display()
```

```python
emp_df.withColumn(
    "joining_date",
    col("joining_date").cast("date")
).display()
```

```python
emp_df.withColumn(
    "emp_id_string",
    col("emp_id").cast("string")
).display()
```

```python
emp_df.withColumn(
    "bonus",
    (col("salary") * 0.10).cast("double")
).display()
```
