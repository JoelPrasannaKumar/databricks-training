### filter() / where()

```python
emp_df.filter(col("salary") > 70000).display()
```

```python
emp_df.filter(col("department") == "IT").display()
```

```python
emp_df.filter(col("age") < 30).display()
```

```python
emp_df.filter(col("gender") == "Female").display()
```

```python
emp_df.filter(col("city") == "Hyderabad").display()
```

### withColumn()

```python
emp_df.withColumn("bonus", col("salary") * 0.10).display()
```

```python
emp_df.withColumn("tax", col("salary") * 0.05).display()
```

```python
emp_df.withColumn("salary_increment", col("salary") + 5000).display()
```

```python
emp_df.withColumn(
    "age_group",
    when(col("age") < 30, "Young")
    .when(col("age") < 35, "Adult")
    .otherwise("Senior")
).display()
```

```python
emp_df.withColumn("yearly_salary", col("salary") * 12).display()
```

### agg()

```python
emp_df.agg(sum("salary").alias("total salary")).display()
```

```python
emp_df.agg(
    avg("age").alias("average_age")
).display()
```

```python
emp_df.agg(max("salary").alias("Max salary"), min("salary").alias("Min salary")).display()
```

```python
emp_df.agg(
    count("*").alias("employee_count")
).display()
```

```python
emp_df.groupBy("department").agg(avg("salary").alias("Average salary"),avg("age").alias("Average age")).display()
```
