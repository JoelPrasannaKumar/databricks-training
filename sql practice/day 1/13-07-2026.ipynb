🟢 Section A: Basic Level (5 Questions)


1️⃣ Display the first 5 rows of the dataset.


```python
%sql
SELECT * FROM Students LIMIT 5;
```

2️⃣ Check the total number of rows and columns in the dataset.


```python
%sql
SELECT (SELECT COUNT(*) FROM Students) AS total_rows, 6 AS total_columns;
```

3️⃣ Find the number of missing values in each column.


```python
%sql
SELECT 
  SUM(CASE WHEN student_id IS NULL THEN 1 ELSE 0 END) as missing_id,
  SUM(CASE WHEN student_name IS NULL THEN 1 ELSE 0 END) as missing_name,
  SUM(CASE WHEN department IS NULL THEN 1 ELSE 0 END) as missing_dept,
  SUM(CASE WHEN city IS NULL THEN 1 ELSE 0 END) as missing_city,
  SUM(CASE WHEN marks IS NULL THEN 1 ELSE 0 END) as missing_marks,
  SUM(CASE WHEN attendance IS NULL THEN 1 ELSE 0 END) as missing_attendance
FROM Students;
```

4️⃣ Replace missing values in the marks column with the average marks.


```python
%sql
WITH AvgMarks AS (SELECT AVG(marks) as avg_m FROM Students WHERE marks IS NOT NULL)
SELECT student_id, student_name, department, city, 
       COALESCE(marks, (SELECT avg_m FROM AvgMarks)) as marks, 
       attendance
FROM Students;
```

🟠 Section C: Column Transformations (5 Questions)


1️⃣1️⃣ Create a new column percentage assuming total marks are 200.


```python
%sql
SELECT *, (marks / 200.0) * 100 AS percentage FROM Students;
```

1️⃣2️⃣ Create a new column grade using:
• A → percentage ≥ 85
• B → percentage ≥ 70
• C → percentage ≥ 50
• Fail → otherwise


```python
%sql
WITH Pct AS (SELECT *, (marks / 200.0) * 100 AS percentage FROM Students)
SELECT *,
  CASE 
    WHEN percentage >= 85 THEN 'A'
    WHEN percentage >= 70 THEN 'B'
    WHEN percentage >= 50 THEN 'C'
    ELSE 'Fail'
  END AS grade
FROM Pct;
```

1️⃣3️⃣ Add 5 bonus marks to every student.


```python
%sql
SELECT *, marks + 5 AS bonus_marks FROM Students;
```

1️⃣4️⃣ Create a new column result where:
• Pass → percentage ≥ 40
• Fail → otherwise


```python
%sql
WITH Pct AS (SELECT *, (marks / 200.0) * 100 AS percentage FROM Students)
SELECT *,
  CASE WHEN percentage >= 40 THEN 'Pass' ELSE 'Fail' END AS result
FROM Pct;
```

1️⃣5️⃣ Convert all student names to uppercase.


```python
%sql
SELECT *, UPPER(student_name) AS student_name_upper FROM Students;
```

🔵 Section D: GroupBy & Aggregation (5 Questions)


1️⃣6️⃣ Find the average marks of each department.


```python
%sql
SELECT department, AVG(marks) as avg_marks FROM Students GROUP BY department;
```

1️⃣7️⃣ Find the maximum marks scored in each city.


```python
%sql
SELECT city, MAX(marks) as max_marks FROM Students GROUP BY city;
```

1️⃣8️⃣ Count the number of students in each department.


```python
%sql
SELECT department, COUNT(*) as student_count FROM Students GROUP BY department;
```

1️⃣9️⃣ Find the department with the highest average marks.


```python
%sql
SELECT department, AVG(marks) as avg_marks FROM Students GROUP BY department ORDER BY avg_marks DESC LIMIT 1;
```

2️⃣0️⃣ Create a pivot table showing average marks by department and city.


```python
%sql
SELECT department,
  AVG(CASE WHEN city = 'Hyderabad' THEN marks END) as Hyderabad,
  AVG(CASE WHEN city = 'Bangalore' THEN marks END) as Bangalore,
  AVG(CASE WHEN city = 'Chennai' THEN marks END) as Chennai,
  AVG(CASE WHEN city = 'Pune' THEN marks END) as Pune
FROM Students GROUP BY department;
```

2️⃣1️⃣ Rank students based on marks (highest marks = Rank 1).


```python
%sql
SELECT *, RANK() OVER (ORDER BY marks DESC) as rank FROM Students WHERE marks IS NOT NULL;
```

2️⃣2️⃣ Find students whose marks are above the overall average marks.


```python
%sql
SELECT * FROM Students WHERE marks > (SELECT AVG(marks) FROM Students);
```

2️⃣3️⃣ Standardize the marks column using Min-Max Scaling.


```python
%sql
WITH MinMax AS (SELECT MIN(marks) as min_m, MAX(marks) as max_m FROM Students)
SELECT *, (marks - min_m) / (max_m - min_m) as scaled_marks
FROM Students, MinMax;
```

2️⃣4️⃣ Categorize attendance into:
• Excellent (≥ 90)
• Good (75-89)
• Average (< 75)


```python
%sql
SELECT *,
  CASE 
    WHEN attendance >= 90 THEN 'Excellent'
    WHEN attendance >= 75 THEN 'Good'
    ELSE 'Average'
  END as attendance_category
FROM Students;
```
